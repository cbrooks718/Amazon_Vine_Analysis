```python
# Read CSV File into dataframe
import pandas as pd
df = pd.read_csv("Vine_table.csv")
df.head()
```

```python
# Filter dataframe to show total_votes greater than or equal to 20
df1 =df[df['total_votes']>=20]
df1.head(20)
```

```python
# Filter dataframe so that the helpful votes divided by total_votes is greater than or equal to 50%
df2 = df1[(df1["helpful_votes"])/(df1["total_votes"])>=0.5]
df2.head(20)
```

```python
# Create new dataframe for paid purchases
paid_df = df2[df2["vine"]=='Y']
paid_df.head()
```

```python
# Create new dataframe for unpaid purchases
unpaid_df = df2[df2["vine"]=='N']
unpaid_df.head()
```

```python
# Total number of reviews, paid
paid_reviews = paid_df['review_id'].count()
paid_reviews
```

```python
# Total number of five-star reviews, paid
paid_five_star = paid_df['star_rating'][paid_df['star_rating']==5.0].count()
paid_five_star
```

```python
paid_five_star_percent = (paid_five_star/paid_reviews)*100
paid_five_star_percent
```

```python
# Total number of reviews, unpaid
unpaid_reviews = unpaid_df['review_id'].count()
unpaid_reviews
```

```python
unpaid_five_star = unpaid_df['star_rating'][unpaid_df['star_rating']==5.0].count()
unpaid_five_star
```

```python
unpaid_five_star_percent = (unpaid_five_star/unpaid_reviews)*100
unpaid_five_star_percent
```
